# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import image

from sklearn.datasets import fetch_openml  # 오픈 소스로 공개된 ML 데이터 셋을 다운로드 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_validate
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Data

```python
mnist = fetch_openml(name='mnist_784')
```

```python
mnist.keys()
```

```python
X = mnist['data']  # data
y = mnist['target'].astype(np.int8)   # target. 문자열을 숫자열로 변환
```

```python
X.shape, y.shape
```

```python
plt.imshow(X[0].reshape((28, 28)), cmap='binary')
plt.show()
```

# Train/Test split

```python
num_train = 60_000
X_train, X_test = X[:num_train], X[num_train:]
y_train, y_test = y[:num_train], y[num_train:]
```

# SGDClassifier

```python
sgd_clf = SGDClassifier(n_jobs=-1, random_state=42)   # 모델 생성
```

```python
sgd_clf.fit(X_train, y_train)   # 모델 훈련
```

```python
sgd_clf.score(X_train, y_train)
```

# KNeighborsClassifier

```python
knn_clf = KNeighborsClassifier(n_jobs=-1)
```

```python
knn_clf.fit(X_train, y_train)
```

```python
# 샘플 100개만 예측
pred = knn_clf.predict(X_train[:100])
```
