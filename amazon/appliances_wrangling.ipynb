```python
# YIELD demo
# A Python program to generate squares from 1
# to 100 using yield and therefore generator
  
# An infinite generator function that prints
# next square number. It starts with 1
def nextSquare():
    i = 1;
  
    # An Infinite loop to generate squares 
    while True:
        yield i*i                
        i += 1  # Next execution resumes 
                # from this point     
  
# Driver code to test above generator 
# function
for num in nextSquare():
    if num > 100:
         break    
    print(num)
```

```python
# metadata import
def parse(path):
    g = gzip.open(path, 'rb') # gzip opens path, in 'read binary' mode
    for l in g:
        yield eval(l) # eval RUNS the code within the parentheses; e.g. eval('1 + 1') returns 2 on the console
        # yield can produce a SEQUENCE of values, rather than sending ONE back to the caller; 
        # We should use yield when we want to iterate over a sequence, but don’t want to store the entire sequence in memory

def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')

dfmeta_gz = getDF('meta_Appliances.json.gz')
dfmeta_gz.head()
```

```python
# drop null values
app_reviews2 = app_reviews.dropna(subset=['title'])
app_reviews.shape, app_reviews2.shape
```

```python
# filling missing reviewerNames...? unnecessary, since reviewerID is alive and well?
# app_reviews2.loc[app_reviews2.reviewerName.isnull(), 'reviewerName'] = app_reviews2.reviewerID.str[0:6]
# I'm not going to use the 'vote' column anyway...
# app_reviews2.loc[app_reviews2.vote.isnull(), 'vote'] = '0'

# filtering out the non-verified purchases
app_reviews2 = app_reviews2[app_reviews2['verified'] == True]

# dropping null values from reviewText, since that's really important...
app_reviews3 = app_reviews2.dropna(subset=['reviewText'])
app_reviews3.isnull().sum()
```

```python
app_reviews_nn = app_reviews3.drop(['image_x', 'image_y', 'style'], axis=1)
app_reviews_nn.isnull().sum()
```

```python
# choosing ONLY the columns that I need...
app_reviews_nn = app_reviews_nn[['Rating', 'reviewTime', 'reviewerID', 'asin', 'reviewText', 'summary', 'unixReviewTime', 'category', 'description', 'also_buy', 'brand', 'main_cat', 'price']]
app_reviews_nn.info()
```

```python
# concatenate-ing review text and summary, then dropping them
app_reviews_nn['review_text'] = app_reviews_nn[['summary', 'reviewText']].apply(lambda x: " ".join(str(y) for y in x if str(y) != 'nan'), axis = 1)
app_reviews_nn = app_reviews_nn.drop(['reviewText', 'summary'], axis=1)
app_reviews_nn.head()
```

```python
# classifying ratings
good_rate = len(app_reviews_nn[app_reviews_nn['Rating'] >= 3])
bad_rate = len(app_reviews_nn[app_reviews_nn['Rating'] < 3])

# Printing rates and their total numbers
print('Good ratings : {} reviews for Appliances'.format(good_rate))
print('Bad ratings : {} reviews for Appliances'.format(bad_rate))
```

```python
# applying new ratings to the Ratings column
app_reviews_nn['rating_class'] = app_reviews_nn['Rating'].apply(lambda x: 'bad' if x < 3 else'good')
app_reviews_nn.head()
```

```python
# checking for duplicates with 'asin'
# app_reviews_nn.duplicated()
# ^ this does not work since some of the columns are lists, which cannot be processed by duplicated()
app_reviews_nn['duplicate'] = app_reviews_nn[['reviewerID', 'asin', 'unixReviewTime', 'review_text']].duplicated()
app_reviews_nn['duplicate'].value_counts()
# DataFrame.duplicated() - checks for duplicate data
# keep=False indicates 'False' for every duplicate entry, rather than the first one or the last one only
# this is acting weirdly... only 44470 entries survive...?
```

```python
app_reviews_nn = app_reviews_nn[app_reviews_nn['duplicate'] == False]
```
